```python
import numpy as np

par = 1000

freq = 2

term_to_maturity = 99

required_yield = 0.001

coup_rate = 20

yields = []

for i in np.arange(0,100,0.1):
    yields.append(25/np.power(10,i))

present_Vals = []

for yld in yields:
    present_val = (coup_rate/100)*par/freq*((1-(1/np.power(1+(yld/(100*freq)),term_to_maturity*freq)))/(yld/(100*freq))) + (par/(np.power(1+(yld/(100*freq)),term_to_maturity*freq)))
    present_Vals.append(present_val)

import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.plot(yields, present_Vals, c="red", alpha=1, label="Price")
```

```python
plt.xscale("log")

plt.xlim(0.00000000001,25)

plt.ylim(-1000,30000)

plt.grid()

plt.xlabel("Yield(%)")

plt.ylabel("Price($)")

plt.title(f"Yield vs Price (Coup., Par, Tot.): {term_to_maturity}Y, Coup {coup_rate}%, \n Par ${par}, semi-annual")

plt.axhline(y = par, xmin = 0, xmax = 25, color='b', linestyle='dashed',alpha=0.5,label=f"Par Value: ${par}")

plt.legend(loc="upper right")
```

```python
plt.savefig("figures/logYield.png")
```
