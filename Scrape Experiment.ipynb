```python
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import pandas as pd
import time
from splinter import Browser





executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

hash_tag = input("enter hashtag: ")

search_url = f"https://twitter.com/search?q=%23{hash_tag}&src=typd&lang=en"

browser.visit(search_url)

num_retweets = []
tweet_text = []

SCROLL_PAUSE_TIME = 1.0

for i in range(1,20):
    page_html = browser.html
    twitter_soup = bs(page_html,"html.parser")

    retweet_data = twitter_soup.find_all("div",class_="stream-item-footer")
    tweet_text_data = twitter_soup.find_all("div",class_="js-tweet-text-container")
    
    uplim = len(retweet_data)
    lowlim = 0
    for j in range(lowlim,uplim):
        retweet_div = retweet_data[j].find("div",class_="ProfileTweet-action ProfileTweet-action--retweet js-toggleState js-toggleRt")
        
        retweet = retweet_div.button.find("span",class_="ProfileTweet-actionCount").text
        retweet2 = retweet.replace('\n','')
        if retweet2 == '':
            retweet3 = '0'
        else:
            retweet3 = retweet2
        num_retweets.append(int(retweet3))
        
        tweet_text_div = tweet_text_data[j]
        tweet_text.append(tweet_text_div.p.text)
        
    lowlim = len(num_retweets)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

results_dict = {
    "retweets":num_retweets,
    "text":tweet_text
}

df = pd.DataFrame(results_dict)
df2 = df.drop_duplicates("text")

sorted_vals = df2.sort_values("retweets", axis=0, ascending=False, inplace=False, kind='quicksort', na_position='last')

top_50 = sorted_vals.head(50)

to_print = top_50["text"].tolist()

for tweet in to_print:
    print(tweet)
    print("--------")
```
