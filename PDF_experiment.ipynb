```python
dfs=[]
rows = len(df.index)
for i in range(0,rows):
    if df.iloc[i,0] == 'T.VALOR':
        top_of_table = i
for i in range(top_of_table + 1,rows):
    if isinstance(df.iloc[i, 0], str) == False:
        if isinstance(df.iloc[i + 1, 0], str) == True:
            obj = {}
            obj["title"] = df.iloc[i,1]
            obj["lowlim"] = i + 1
            obj["uplim"] = 'provisional'
            if len(dfs) > 0:
                dfs[len(dfs) - 1]["uplim"] = i - 1
            dfs.append(obj)
```
