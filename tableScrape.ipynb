```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
```

```python
url = "http://www.banxico.org.mx/apps/stdview.html?url=/apps/DAO-web/4/20/DistribucionTenencia.html"
```

```python
executable_path = {"executable_path":"chromedriver.exe"}
browser = Browser("chrome",**executable_path,headless=False)
```

```python
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = bs(html,"html.parser")
```

```python
table = soup.find("table",attrs={"border":"0","cellpadding":"5","cellspacing":"0","width":"80%","align":"center"})
```

```python
tables = soup.find_all("table",attrs={"border":"0","cellpadding":"5","cellspacing":"0","width":"80%","align":"center"})
```

```python
dict_list = []
for table in tables:
    table_data = table.find_all("tr")[1:]
    CI = []
    fechVenc = []
    repBanx = []
    garBanx = []
    vadBan = []
    Banca = []
    SSI = []
    Aseg = []
    Otros = []
    ResPai = []
    Ext = []
    Total = []
    
    import pdb; pdb.set_trace()
    
    for row in table_data:
        row_data = row.find_all("td")
        CI.append(row_data[0].text)
        fechVenc.append(row_data[1].text)
        repBanx.append(float(row_data[2].text.replace(",","")))
        garBanx.append(float(row_data[3].text.replace(",","")))
        vadBan.append(float(row_data[4].text.replace(",","")))
        Banca.append(float(row_data[5].text.replace(",","")))
        SSI.append(float(row_data[6].text.replace(",","")))
        Aseg.append(float(row_data[7].text.replace(",","")))
        Otros.append(float(row_data[8].text.replace(",","")))
        ResPai.append(float(row_data[9].text.replace(",","")))
        Ext.append(float(row_data[10].text.replace(",","")))
        Total.append(float(row_data[11].text.replace(",","")))
    
        table_dict = {
            "CI":CI,
            "Fecha de Vencimiento":fechVenc,
            "Garantías recibidas por Banxico":repBanx,
            "Valores adquiridos por Banxico":vadBan,
            "Sector Bancario":Banca,
            "Siefores y Sociedades de Inversión":SSI,
            "Aseguradoras y Afianzadoras":Aseg,
            "Otros Residentes en el Pais":ResPai,
            "Residentes en el Extranjero":Ext,
            "Total en Circulación":Total
        }
    
        dict_list.append(table_dict)
    
```
